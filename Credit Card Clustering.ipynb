```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('CC GENERAL.csv')
data.info()
```

```python
data.head()
```

```python
data=data.drop(columns=['CUST_ID'])
```

```python
data.head()
```

```python
data['TENURE'].unique()
```

```python
data['CREDIT_LIMIT'].fillna(value=data['CREDIT_LIMIT'].mean(), inplace=True)
```

```python
data['MINIMUM_PAYMENTS'].fillna(value=0, inplace=True)
```

```python
data.info()
```

```python
cols=[]
for i in data.columns:
    if(abs(data[i].skew())>0.75):
        cols.append(i)
    print(abs(data[i].skew()),'\t',i)
```

Feature Engineering

```python
for i in cols:
    data[i] = np.log1p(data[i])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
data[data.columns] = sc.fit_transform(data[data.columns])
data.head()
```

Implementing KMeans

```python
from sklearn.cluster import KMeans
```

```python
inertia=[]
for i in range(1,20):
    kmeans = KMeans(i)
    kmeans.fit(data[data.columns])
    iner = kmeans.inertia_
    inertia.append(iner)

number_clusters = range(1,20)
plt.plot(number_clusters,inertia)
plt.title('Elbow')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(3)
kmeans.fit(data[data.columns])
```

```python
data['kmeans']=kmeans.predict(data[data.columns])
```

```python
data.head()
```

```python
d0=data.loc[data['kmeans']==0]
d0[d0.columns[:-1]]=sc.inverse_transform(d0[d0.columns[:-1]])

d1=data.loc[data['kmeans']==1]
d1[d1.columns[:-1]]=sc.inverse_transform(d1[d1.columns[:-1]])

d2=data.loc[data['kmeans']==2]
d2[d2.columns[:-1]]=sc.inverse_transform(d2[d2.columns[:-1]])
```

```python
d1
```

```python
d=[d0,d1,d2]
for i in d:
    for j in i.columns[:-1]:
        i[j]=np.exp(i[j])-1
    print('\n')
    print(i.describe().drop(index=['25%','50%','75%','std']).T)
```

Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

ag = AgglomerativeClustering(n_clusters=3, linkage='ward', compute_full_tree=True)
ag = ag.fit(data[data.columns[:-1]])
data['agglom'] = ag.fit_predict(data[data.columns[:-1]])
data.head()
```

```python
list(data['kmeans'].groupby(data['kmeans']).size())
```

```python
list(data['agglom'].groupby(data['agglom']).size())
```

```python
d0=data.loc[data['agglom']==0]
d0[d0.columns[:-2]]=sc.inverse_transform(d0[d0.columns[:-2]])

d1=data.loc[data['agglom']==1]
d1[d1.columns[:-2]]=sc.inverse_transform(d1[d1.columns[:-2]])

d2=data.loc[data['agglom']==2]
d2[d2.columns[:-2]]=sc.inverse_transform(d2[d2.columns[:-2]])
```

```python
d=[d0,d1,d2]
for i in d:
    for j in i.columns[:-1]:
        i[j]=np.exp(i[j])-1
    print('\n')
    print(i.describe().drop(index=['25%','50%','75%','std','min','max']).T)
```

Implementing DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
e=[1,2,3,4,5]
ms=[10,15,20,25,30,35]
df=pd.DataFrame(columns=e,index=ms)
for i in e:
    vals=[]
    for j in ms:
        db=DBSCAN(eps=i,min_samples=j)
        db=db.fit(data[data.columns[:-2]])
        data['db']=db.labels_
        vals.append(len(list(data['db'].groupby(data['db']))))
        data=data.drop(columns=['db'])
    df[i]=vals
for i in df.columns:
    df[i]=df[i]-1
df
```

```python
db=DBSCAN(eps=2,min_samples=20)
db=db.fit(data[data.columns[:-2]])
data['db']=db.labels_
```

```python
do=data.loc[data['db']==-1]
do[do.columns[:-3]]=sc.inverse_transform(do[do.columns[:-3]])

d0=data.loc[data['db']==0]
d0[d0.columns[:-3]]=sc.inverse_transform(d0[d0.columns[:-3]])

d1=data.loc[data['db']==1]
d1[d1.columns[:-3]]=sc.inverse_transform(d1[d1.columns[:-3]])

d2=data.loc[data['db']==2]
d2[d2.columns[:-3]]=sc.inverse_transform(d2[d2.columns[:-3]])
```

```python
d=[do,d0,d1,d2]
for i in d:
    for j in i.columns[:-1]:
        i[j]=np.exp(i[j])-1
    print('\n')
    print(i.describe().drop(index=['25%','50%','75%','std','min','max']).T)
```
