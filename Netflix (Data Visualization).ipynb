```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d=pd.read_csv('netflix_titles.csv')
d.info()
d.head()
```

```python
d=d.drop(['show_id','description'],axis=1)
d
```

```python
d.dropna(subset = ['duration','rating','country','date_added'], inplace=True)
d.info()
d.head()
```

```python
d['date_added'] = d['date_added'].str.strip()
d['date_added'] = pd.to_datetime(d['date_added'], format='%B %d, %Y')
d.info()
```

```python
def con(s):
    ss=0
    for i in s:
        if (i>='0' and i<='9'):
            ss=ss*10+int(i)
        else: break
    return ss
d['duration']=list(map(lambda x : con(str(x)), d['duration']))
d.head()
```

```python
d.country.value_counts()
```

```python
d_movie=d.loc[(d['type']=='Movie')]
d_show=d.loc[(d['type']=='TV Show')]
d_movie.info()
d_show.info()
```
